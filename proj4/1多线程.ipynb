# 多线程
线程和进程：每一个应用在运行的时候占用内存的一部分，这部分就叫做进程，每一个进程下有若干个线程。进程是**资源单位**，线程是**执行单位**。一个进程中至少有一个线程，线程是进程得以运行的保证。就好比世界上有很多正在进行的公司（进程），每个公司要想运行起来，里面必须要有工作的人（线程），多个人做自己的业务，总效率就会提升，这就是多线程。启动每一个程序都有一个默认的主线程。

在python里面想实现多线程，需要用到python的`threading`模块。

```python
from threading import Thread
def func():
    for i in range(100):
        print("func",i)
```

```python

t=Thread(target=func)   #创建线程并安排任务
t.start()               #给线程状态设置为“可以开始工作了”状态，此时t将在之后自行开始工作，具体开始时间由CPU决定
#同时，如果CPU忙，可能会出现输出跟下面混合输出的问题。
for i in range(100):
    print('main',i)
```

实现多线程的另一个方法：类

```python
class mythread(Thread):         #这里指mythread是Thread的一个子类，所以mythread也是一个线程
    def run(self) :             #这个函数是固定的,线程被执行的时候，被执行的就是run()
        for i in range(100):
            print("子线程",i)
t=mythread()
# t.run()                       #这个不行，实际上还是一个单线程
t.start()                       #这个才是多线程
for i in range(100):
    print('main',i)
```

```python
def func1(name):
    for i in range(100):
        print(name,i)
t1=Thread(target=func1,args=("线程甲",)) #args用来传参的时候，规定了必须传入元组，即便只有一个参数也得在后面加上逗号。
t2=Thread(target=func1,args=("线程乙",))
t1.start()
t2.start()
```
