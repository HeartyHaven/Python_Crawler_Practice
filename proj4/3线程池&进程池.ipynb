# 利用多线程完成大规模数据爬取

线程池：一次性开辟一些线程，用户直接给线程池提交任务。比如一个有50个线程的线程池，给了10000个任务，那么调度都由线程池完成，我们不用管里面是怎么做的。进程池同理。

```python
from concurrent.futures import ThreadPoolExecutor,ProcessPoolExecutor   #创建线程池和进程池
def fn(name):
    for i in range(100):
        print(name,i)
if __name__=='__main__':
    with ThreadPoolExecutor(50) as t:       #开启一个有50个线程的线程池
        for i in range(100):
            t.submit(fn,name=f"线程{i}")     #提交给线程池100个任务，任务的内容是根据参数执行fn
    print('123')                            #with外面的内容在等待进程池全部结束之后再执行，这就是守护线程。
```

进程池也是这个语法，只不过是把ThreadPoolExecutor换成了ProcessPoolExecutor。
