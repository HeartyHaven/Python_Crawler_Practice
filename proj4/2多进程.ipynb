# 多进程

```python
from multiprocessing import Process

def func():
    print("子进程开始执行")
    try:
        for i in range(100): 
            print("子进程", i)
    except Exception as e:
        print(f"子进程出错:{e}")

p = Process(target=func)
p.start()
p.join()

for i in range(100):      
    print("主进程", i)                  #我的电脑占用资源过多，所以可能没法成功开启多进程。
```

python这里虽然多线程跟多进程开启的方法几乎完全一样，但是实际背后的工作完全不同，所以python极大地简化了开发者的工作。
