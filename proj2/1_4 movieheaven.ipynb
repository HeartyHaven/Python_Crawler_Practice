# 获取电影天堂首页的全部数据

## 任务：
- 定位到2022必看电影
- 从2022必看电影中提取到子页面的链接地址
- 请求子页面的链接地址，得到下载地址

```python
import requests
domain="https://www.dytt8.net/"
resp=requests.get(domain,verify=False)#verify这个是为了解决一个ssl error，有时出现（如果网站有加密），有时不会出现
resp.encoding="gb2312"#指定字符集，从网页源代码中看
print(resp.text)
```

拿到ul里面的超链接和相应的电影

```python
import re
obj1=re.compile("最新电影更新.*?<ul>(?P<ul>.*?)</ul>",re.S)
obj2=re.compile(r"<a href='(?P<urls>.*?)'>(?P<names>.*?)</a>",re.S)
childhref_list=[]
result=obj1.finditer(resp.text)
for it in result:
    ul=it.group("ul")
    # print(ul)
    result2=obj2.finditer(ul)
    # print(result2)
    for itt in result2:
        # print(itt.group())
        childhref=domain+itt.group('urls').strip("/")#拼接子页面的链接：域名+子页面地址，是否这么做需要自己看
        childhref_list.append(childhref)
        print(itt.group('urls'))
        print(itt.group('names'))
```

提取子页面内容


```python
obj3=re.compile(r'电影下载地址2：<a href="(?P<download>.*?)" target.*?>',re.S)
for href in childhref_list[1:]:
    childresp=requests.get(href)
    childresp.encoding='gb2312'
    # print(childresp.text)#test
    findherf=obj3.search(childresp.text)
    print(findherf.group('download'))
```
