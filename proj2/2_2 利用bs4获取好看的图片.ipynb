# 获取壁纸
操作步骤：

- 拿到主页面的源代码，然后提取到子页面的链接地址href；
- 通过这个href拿到子页面的内容，从子页面中提取下载地址。
- 通过下载地址下载图片。

```python
from bs4 import BeautifulSoup
import re
import requests
```

获取网页信息

```python
url="https://www.umei.cc/bizhitupian/"
resp=requests.get(url)
resp.encoding="utf-8"
print(resp.text)
```

内容交给BeautifulSoup拿到子页面的链接

```python
import time
main_page=BeautifulSoup(resp.text,"html.parser")
res=main_page.find('div',class_='taotu-main')
# print(res)
plist=res.find_all('p')
# print(plist)
obj=re.compile(r'<p><a href="(?P<href>.*?)"',re.S)
# <p><a href="/bizhitupian/diannaobizhi/321667.htm" title="iPhone，摄影，桥，纽约">iPhone，摄影，桥，纽约</a></p>

for p  in plist:
    a=p.find('a')
    # print(a)
    href=a.get('href')#用get获取标签的属性值
    #拿到子页面的源代码
    # print ("https://www.umei.cc"+href)
    child_page_resp=requests.get("https://www.umei.cc"+href)
    child_page_resp.encoding='utf-8'
    # print(child_page_resp.text)
    child_page_text=BeautifulSoup(child_page_resp.text,"html.parser")
    #从子页面中获取下载路径
    child_page_target=child_page_text.find('div',class_="big-pic")
    img=child_page_target.find('img')
    # print(child_page_target)
    src=img.get('src')
    # print(img.get('src'))
    # break
    """
    
    下载图片
    ↓ ↓ ↓
    """
    img_resp=requests.get(src)
    imgname=src.split("/")[-1] #取最后一个/以后的所有内容作为文件名
    with open("img/testimgs/"+imgname,"wb") as f:
        f.write(img_resp.content) #这里拿到的是字节内容，不是直接一个图片,通过写入将图片内容保存到图片
    f.close
    print("over in "+imgname)
    time.sleep(1)
print("all over!")
```

项目内的文件夹产生了更新，pycharm会对更新的内容生成索引。所以当内容特别特别多的时候，pycharm生成索引的速度就会越来越慢，整体的效率也就大大降低。为了解决这个问题，可以把存放图片的文件夹设置为excluded形式，这样在该文件夹进行更新的时候就不再生成索引。
