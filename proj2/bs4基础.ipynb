方案：

```python
! pip install bs4
# import bs4
from bs4 import BeautifulSoup
import requests
```

```python
# url="http://www.xinfadi.com.cn/getCat.html"#因为这个网址改成客户端装配了，所以没法体现bs4的功能，这里换一个目标网址。
url="http://www.xinfadi.com.cn/index.html"
headers={
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36"
}
resp=requests.get(url)
print(resp.text)
```

解析数据：

```python
#把源代码交给beautifulsoup进行处理，返回一个bs4对象
page=BeautifulSoup(resp.text,"html.parser")
#不加后面那段话bs4不知道你给的是什么类型的语言，默认用html解析，但会警告。
```

查找数据：

bs4里面只有2个方法：
1. `find(标签，属性=值)` 返回第一个符合条件的内容；
2. `find_all(标签，属性=值)`  返回所有符合条件的内容。

```python
div=page.find('div',class_="news")
#为了防止html中的class跟py中的class冲突，bs4引入class_代替class。
div=page.find('div',attrs={"class":"news"})
#这样写也可以，跟上面效果完全相同，但好像看起来规范一些，而且可以避免class
# print(div)
news=div.find_all("span")
print(news)
```
